```python
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression as lrg
```

```python
file = "csv/01 revenue.csv"
data = pd.read_csv(file)
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns=["production_budget_usd"])
y = pd.DataFrame(data, columns=["worldwide_gross_usd"])
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X, y, alpha=.1)
plt.title('Film Cost and Revenue')
plt.xlabel('Product Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0, 5e8)
plt.ylim(0, 3e9)
plt.show()
```

```python
regression = lrg()
regression.fit(X, y)
```

Slope Coefficient(ضریب شیب):

```python
regression.coef_
```

Intercept(مقدار ثابتمون):

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X, y, alpha=.1)
plt.plot(X, regression.predict(X), color='red', linewidth=5)
plt.title('Film Cost and Revenue')
plt.xlabel('Product Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0, 5e8)
plt.ylim(0, 3e9)
plt.show()
```

R squared:

```python
regression.score(X, y)
```
