```python
import pandas as pd
from sklearn.linear_model import LinearRegression as lrg
from matplotlib import pyplot as plt
```

```python
file = "csv/imdb.csv"
data = pd.read_csv(file)
```

```python
ratio = pd.DataFrame(data, columns=['Production to Worldwide Ratio'])
imdb = pd.DataFrame(data, columns=['IMDB'])
```

```python
reg = lrg()
reg.fit(imdb, ratio)
```

```python
plt.scatter(imdb, ratio, alpha=.5)
plt.plot(imdb, reg.predict(imdb), color="red", linewidth=1)
plt.plot(imdb, reg.predict(imdb)+.25, color="red", alpha=.5)
plt.plot(imdb, reg.predict(imdb)-.25, color="red", alpha=.5)

plt.xlabel('IMDB Rate')
plt.ylabel('Production to Worldwide Ratio')
```
