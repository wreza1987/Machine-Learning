```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## define $f(x)$

```python
def f(x):
    return x**3 - x + 1
```

## define $df$ at any point

```python
def df(x):
    return 3*x**2 - 1
```

```python
x = np.linspace(-1.5,1.5,100)
```

```python
plt.figure(figsize=(6,7))
plt.plot(x, f(x), color="blue")
plt.plot(x, df(x), color="red")
plt.grid()
plt.show()
```

```python
x1 = 1
x2 = 0
eps = .1
Xlist = [x1]
for i in range(30):
    x2 = x1
    x1 = x2 - eps*df(x2)
    Xlist.append(x1)
print("the answer: ", x2)
Xnp = np.array(Xlist)
```

```python
plt.figure(figsize=(6,7))
plt.plot(x, f(x), color="darkblue", alpha=.6)
plt.scatter(Xlist, f(Xnp), color="red", alpha=.4, s=50)
plt.plot(x, df(x), color="darkred", alpha=.6)
plt.scatter(Xlist, df(Xnp), color="blue", alpha=.4, s=50)
plt.grid()
plt.show()
```

## $$ g(x) = x^4 - 4x^2 + 5$$

```python
x_list = np.linspace(-2, 2, 1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.plot(x_list, g(x_list), color="blue")
plt.plot(x_list, dg(x_list), color="red")
plt.grid()
```

```python
x1 = -2
x2 = -1
eps = 1e-6
xlist = []
for i in range(100):
    x0 = (x1+x2)/2
    xlist.append(x0)
    if abs(dg(x0))>eps:
        if dg(x0)*dg(x1) < 0:
            x2 = x0
        else:
            x1 = x0
    else:
        break
print("number of loops: ", i)
print("the answer: ", x0)
```

```python
xlist = np.array(xlist)
plt.figure(figsize=(15, 10))
plt.subplot(2,2,1)
plt.title("approch on function")
plt.xlim(-1.55,-1.2)
plt.ylim(0.95, 1.35)
plt.plot(x_list, g(x_list), color="blue", alpha=.6)
plt.scatter(xlist, g(xlist), color="darkred", alpha=.4)

plt.subplot(2,2,2)
plt.title("approch on derivative")
plt.xlim(-1.55,-1.2)
plt.ylim(-3, 3)
plt.plot(x_list, dg(x_list), color="red", alpha=.6)
plt.scatter(xlist, dg(xlist), color="darkblue", alpha=.4)
plt.grid()

num = np.array(range(len(xlist)))

plt.subplot(2,2,3)
plt.title("learning rate of function")
plt.ylim(0.95, 1.05)
plt.plot(num, g(xlist), color="blue", alpha=.6)
plt.scatter(num, g(xlist), color="darkred", alpha=.4)

plt.subplot(2,2,4)
plt.title("learning rate of derivative")
plt.ylim(-.5, .5)
plt.plot(num, dg(xlist), color="red", alpha=.6)
plt.scatter(num, dg(xlist), color="darkblue", alpha=.4)
plt.grid()
```

```python
x1 = -2
x2 = -1
eps = 1e-12
def ddg(x):
    return 12*x**2 - 8
xlist = []
for i in range(100):
    x0 = x1 - dg(x1)/ddg(x1)
    xlist.append(x0)
    if abs(dg(x0))>eps:
        x1 = x0
    else:
        break

print("number of loops: ", i)
print("the answer: ", x0)
```

```python
xlist = np.array(xlist)
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
plt.xlim(-1.55,-1.2)
plt.ylim(0.95, 1.35)
plt.plot(x_list, g(x_list), color="blue", alpha=.6)
plt.scatter(xlist, g(xlist), color="darkred", alpha=.4)
plt.subplot(1,2,2)
plt.xlim(-1.55,-1.2)
plt.ylim(-3, 3)
plt.plot(x_list, dg(x_list), color="red", alpha=.6)
plt.scatter(xlist, dg(xlist), color="darkblue", alpha=.4)
plt.grid()
```
